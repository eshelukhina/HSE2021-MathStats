```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import factorial
import math
```

```python
def get_std_uniform(k, theta):
    deviation = np.zeros(k)
    n = 10000
    for _ in range(100): # num of iterations
        sample = np.random.uniform(0, theta, n)
        for cur_k in range(1, k+1):
            calc_theta = ((np.sum(sample ** cur_k) / n) * (cur_k + 1)) ** (1.0 / cur_k)
            deviation[cur_k - 1] += ((calc_theta - theta) ** 2) / 100
    return np.sqrt(deviation)
```

```python
def get_std_exponential(k, theta):
    deviation = np.zeros(k)
    n = 10000
    for _ in range(100): # num of iterations
        sample = np.random.exponential(1.0 / theta, n)
        for cur_k in range(1, k+1):
            calc_theta = ((np.sum(sample ** cur_k) / n) / math.factorial(cur_k)) ** (1.0 / cur_k)
            deviation[cur_k - 1] += ((calc_theta - theta) ** 2) / 100
    return np.sqrt(deviation)
```

```python
def plot_std_uniform(k, theta):
    x = np.arange(0, k, 1)
    y = get_std_uniform(k, theta)
    plt.plot(x, y)
    plt.show()
```

```python
def plot_std_exponential(k, theta):
    x = np.arange(0, k, 1)
    y = get_std_exponential(k, theta)
    plt.plot(x, y)
    plt.show()
```

```python
plot_std_uniform(100,2)
```

```python
plot_std_exponential(100,2)
```
